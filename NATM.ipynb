# News Articles Text Mining

```python
import os

path = 'E:\\NLP data\\Dats sets\\mini_newsgroups\\mini_newsgroups\\sci.space'

#Reading the multiple files from folder altogether using for loop and storing into doc_dict document dictionary

doc_dict = {}

for subdir,dirs,files in os.walk(path):
    for file in files:
        file_path = subdir + os.path.sep + file 
        f = open(file_path, 'rb')
        text = str(f.read())
        doc_dict[file] = text 
```

```python
#Performing lowercase conversion on all the strings in the dictionary
for my_var in doc_dict:
    doc_dict[my_var] = doc_dict[my_var].lower()
doc_dict['61256']
```

```python
#Next is removing unnecessary numbers
import re
for my_var in doc_dict:
    doc_dict[my_var] = re.sub(r'\d+',           # The pattern to search for (d+ :digits which can repeat unlimitedly)
                      " ",                   # The pattern to replace it with (" " : blank space)
                      doc_dict[my_var])  # The text to search
doc_dict['61256']
```

```python
#Removing Punctuations
import re
for my_var in doc_dict:
    doc_dict[my_var] = re.sub(r'\W+|\_',           # The pattern to search for (symbols like punctuations or underscore)
                      " ",                   # The pattern to replace it with
                      doc_dict[my_var])  # The text to search
doc_dict['61256']
```

```python
#Removing general english stop words
from nltk.corpus import stopwords
from nltk import word_tokenize
stop = stopwords.words('english')
for my_var in doc_dict:
    doc_dict[my_var] = ' '.join([i for i in word_tokenize(doc_dict[my_var]) if i not in stop])
doc_dict['61256']
```

```python
#Removing Custom Words
custom_stop = stopwords.words('english') + ['news', 'writes', 'told', 'n']
for my_var in doc_dict:
    doc_dict[my_var] = ' '.join([i for i in word_tokenize(doc_dict[my_var]) if i not in custom_stop])
doc_dict['61256']
```

```python
#Lemmatizing
from nltk.stem import WordNetLemmatizer 
lemmatizer = WordNetLemmatizer()               #Choosing the Lemmatizer
for my_var in doc_dict:
    doc_dict[my_var] = ' '.join([lemmatizer.lemmatize(i) for i in word_tokenize(doc_dict[my_var])])
doc_dict['61256']
```

```python
#Stemming
doc_dict_1=doc_dict
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
for my_var in doc_dict_1:
    doc_dict_1[my_var] = ' '.join([stemmer.stem(i) for i in word_tokenize(doc_dict_1[my_var])])
doc_dict_1['60804']
```
